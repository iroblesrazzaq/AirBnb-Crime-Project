```python
def find_rows_within_radius(df, lon, lat, radius):
    # Create a KDTree from the longitude and latitude columns
    tree = KDTree(df[['longitude', 'latitude']])

    # Find the indices of all points within the specified radius
    indices = tree.query_radius([[lon, lat]], r=radius/6371.0)[0]

    # Return the rows corresponding to the found indices
    return df.iloc[indices]
```
